# Scikit-Lean + Graphviz Demo

This notebook is a simple example of visualizing a Scikit-Learn `DecisionTreeClassifier` in a Jupyter notebook using [`graphviz`](https://graphviz.readthedocs.io/en/stable/index.html). Please see the Graphviz [downloads](https://graphviz.readthedocs.io/en/stable/manual.html?highlight=read#installation) page; you need to install the `graphviz` binary in addition to the Python bindings given by the `graphviz` pip package.

```python
from pprint import pprint

from graphviz import Source
from sklearn.datasets import load_iris
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)

model = DecisionTreeClassifier(max_depth=4).fit(X_train, y_train)
pprint(classification_report(y_test, model.predict(X_test)))
```

This visualization isn't particularly interesting because the features aren't named here. I encourage you to dig into the Graphviz documentation (linked above) to customize the vizualization.

```python
path = 'tree.dot'
export_graphviz(model, out_file=path)
Source.from_file(path)
```
